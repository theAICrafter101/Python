# **Boolean and Conditionals**

- Conditional statements let you control the flow of your program based on whether certain conditions are true or false.

### Comparison Operators: 
- operators that let you compare two or more values, and return a boolean value.

- **Equal (==)**: checks if two vlaues are equal
- **Not Equal (!=)**: checks if two values are not equal
- **Greater than (>)**: checks if value on the left is greater than the value on the right.
- **Less than (<)**: checks if value on the left is less than the value on the right.
- **Greater than or equal (>=)**: checks if value on left is greater than or equal to the value on the right.
- **Less than or equal (<=)**: checks if the value on the left is less than or equal to value on the right.

```python
print(3 > 4) # False
print(3 < 4) # True
print(3 == 4) # False
print(4 == 4) # True
print(3 != 4) # True
print(3 >= 4) # False
print(3 <= 4) # True
```

## If statement

- The most basic conditional in Python is **if** statement.

Basic syntax:
- start with **if** keyword.
- **a condition** that evaluates to True or False.
- followed by a colon **(:)**
- indent the next line after condition to define a code block for if statement

```python
if condition:
    pass # Code to be executed if condition is true
         # Pass: Placeholder for future code
```

```python
age = 18
if age >= 18:
    print("You are an adult.") # Output: You are an adult.
```

- If you don't indent code after if condition, it will throw an error.

```python
age = 18
if age >= 18:
print("You are an adult.") # IndentationError: expected an indented block
```

## Else statement

- If the **if** condition is false nothing will be printed
- This is where you can use **else** statement

```python
# Basic if-else syntax
if condition:
    # Code to be executed if condition is true
else:
    # Code to be executed if condition is false
```

```python
age = 12
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.") # Output: You are a minor.
```

## Elif statement

- If you want to account for multiple conditions, you will be using **elif** for this.

```python
# Basic syntax for if-elif-else
if condition1:
    # Code to be executed if condition1 is true
elif condition2:
    # Code to be executed if condition2 is true
else:
    # Code to be executed if both conditions are false
```

```python
age = 12
if age >= 18:
    print("You are an adult.")
elif age >= 13:
    print("You are a teenager.")
else:
    print("You are a child.") # Output: You are a child.
```

```python
age = 2
if age >= 65:
    print("You are a senior citizen.")
elif age >= 30:
    print("You are an adult.")
elif age >= 18:
    print("You are a young adult.")
elif age >= 13:
    print("You are a teenager.")
elif age >= 3:
    print("You are a child.")
else:
    print("You are a toddler.") # Output: You are a toddler.
```

### Nested Conditional Statements

```python
is_citizen = True
age = 25

if is_citizen:
    if age >= 18:
        print("You are eligible to vote.") # Output: You are eligible to vote.
else:
    print("You are not eligible to vote due to age.")
```

## Logical Operators

These are the logical operators in Python:
- and
- or
- not

These values evaluate to True:
- non-zero numbers
- non-empty strings

These values always evaluate to False:
- None
- False
- Integer 0
- Float 0.0
- Empty strings ""

- **and**: takes two operands and returns the first operand if it is falsy, otherwise, it returns the second operand.
- also known as short-circuit operator

```python
is_citizen = True
age = 20

print(is_citizen and age) # 20
```

```python
is_citizen = True
age = 20

if is_citizen and age >= 18:
    print("You are eligible to vote.") # Output: You are eligible to vote.
else:
    print("You are not eligible to vote.")
```

- **or**: returns the first operand if it is truthy, otherwise, it returns the second operand
- returns True if at least one operand is true.

```python
age = 19
is_employed = False

print(age or is_employed) # 19
```

```python
age = 19
is_student = True

if age < 18 or is_student:
    print("You are eligible for a student discount.") # Output: You are eligible for a student discount.
else:
    print("You are not eligible for a student discount.")
```

- **not**: takes a single output and inverts its boolean value.

```python
print(not '') # True, because empty string is falsy
print(not 'Hello') # False, because non-empty string is truthy
print(not 0) # True, because 0 is falsy
print(not 1) # False, because 1 is truthy
print(not False) # True, because False is falsy
print(not True) # False, because True is truthy
```

```python
is_admin = False

if not is_admin:
    print("Access denied. Admins only.") # Output: Access denied. Admins only.
else:
    print("Welcome, Admin!")
```
