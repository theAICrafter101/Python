### Common data types in Python.

Data type describes the kind of value a variable holds.
- You do not need to explicitly tell Python the types for variables, it already knows it.

```python
name = "John Doe" # Python knows this is a string.
age = 20 # Python knows this is an integer.
```

A drawback:
- Since Python is a dynamically typed language, the bugs and errors are detected only when a program runs, not when the program compiles.

Here are the most common data types used in Python.
- You can use type() function to get the data type of a variable.

Integers: A whole number without decimals.

```python
my_integer_var = 20
print(type(my_integer_var))  # Output: <class 'int'>
```

Float: A number with a decimal point.

```python
my_float_var = 20.5
print(type(my_float_var))  # Output: <class 'float'>
```

String: A sequence of characters enclosed in single or double quotation marks.

```python
my_string_var = "Hello, World!"
print(type(my_string_var))  # Output: <class 'str'>
```

Boolean: A true or false type.

```python
my_boolean_var = True
print(type(my_boolean_var))  # Output: <class 'bool'>
```

Set: An unordered collection of unique elements.

```python
my_set_var = {1, 2, 3}
print(type(my_set_var))  # Output: <class 'set'>
```

Dictionary: A collection of key-value pairs enclosed in curly braces.

```python
my_dictionary_var = {"name": "Nain", "age": 20}
print(type(my_dictionary_var))  # Output: <class 'dict'>
```

Tuple: An immutable ordered collection, enclosed in brackets.

```python
my_tuple_var = (1, 9, 6)
print(type(my_tuple_var))  # Output: <class 'tuple'>
```

Range: A sequence of numbers, often used in loops.

```python
my_range_var = range(6)
print(type(my_range_var))  # Output: <class 'range'>
```

List: An ordered collection of elements that supports different data types.

```python
my_list = [22, 'Hello world', 9.8, False]
print(type(my_list))  # Output: <class 'list'>
```

None: A special value that represents the absence of a value.

```python
my_none_var = None
print(type(my_none_var))  # Output: <class 'NoneType'>
```

The built-in isinstance() function lets you check if a variable matches a specific data type.

- isinstance('Hello world', str) # True
- isinstance(True, bool) # True
- isinstance(42, int) # True
- isinstance('John Doe', int) # False
