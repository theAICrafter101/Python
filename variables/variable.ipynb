# Variables
- Variables are like containers used for storing and referencing data of different types.
- This how you can declare a variable:
1. write name of variable on the left.
2. Then put an assignment operator (=).
3. Finally, the value you want to assign.

```python
name = 'John Doe'
age = 25
```

## Rules for naming the variables
Variable names:
- can only start with a letter, or an underscore (_), not a number.
- can only contain alphanumeric characters (a-z), (A-Z), (0-9), and underscore.
- are case-sensitive, e.g., age, Age, AGE are all considered unique.
- can not be one of Python's reserved keywords such as if, class, or def.

#### Best Practices to name the variables
1. names should be in lowercase, with separate words separated by an underscore (snack_case).

```python
my_varaible_name = "Python Programming"
```

2. Use descriptive names for vaiables.

```python
user_age = 20
```

3. Avoid using single letter names.

```python
x = 56 # What is meant by x here?
```
