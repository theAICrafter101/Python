### How does the Print function works?

In Python, you can use "print" function to display the output to the terminal.

```python
print("Hello, World!")
```

The string -the sequence of characters surrounded by either single (') or double (") quotes.

In print("Hello, World!"), the string "Hello, World!" is an argument passed to the "print" function.

```python
print("My favorite colors are", "blue", "red", "and", "white.")
```

Python automatically adds a space between each item when they are separated by a comma.
